* ## **Exploring Convolutional Neural Networks, Data Augmentation, and Transfer Learning**



> ### **Convolutional Neural Networks (CNNs)**

In the vast and ever-evolving world of machine learning, Convolutional Neural Networks (CNNs) stand out as a groundbreaking force, particularly in handling image data. Their design is a nod to the human visual cortex, tailored to recognize patterns directly from pixel data with minimal preprocessing.

**Layers in CNNs contain three types of layers:**
* **Convolutional Layers:** To produce feature maps, these layers apply filters to the input data and carry out a convolution operation. These filters aid the network in identifying particular characteristics like textures, edges, or more intricate patterns in deeper layers.

* **Pooling Layers:** Convolutional layers are followed by pooling layers, which decrease the input volume's breadth and height for the subsequent convolutional layer. By lowering the number of parameters, memory use, and computational strain, they lessen the chance of overfitting. Max pooling, which uses the maximum value in each feature map window, is the most popular method of pooling.

* **Fully Connected Layers:** At the end of the network, fully connected layers identify the class of the input image by utilizing the features that were extracted by the pooling and convolutional layers.





> ### **Data Augmentation**

An approach called "data augmentation" makes more diverse data available for model training without having to gather additional data. Data augmentation approaches, as used in image processing and CNNs, entail modifying the current images in a dataset to produce newly altered versions of these images. 

**Common techniques include the following:**

* **Random Erasing:** randomly removing part of the images.

* **Noise Injection:** Adding random noise to images.

* **Color Space Transformations:** Adjusting brightness, contrast, or saturation.

* **Geometric Transformations:** Such as rotating, flipping, scaling, and cropping.

Data augmentation improves the model's ability to generalize to new, unseen data and helps avoid overfitting. It's very helpful when there is a small dataset available.



> ### **Transfer Learning**

Transfer learning is a machine learning technique in which a model created for one job is used for another task as the foundation for a new model. Because it can train deep neural networks with relatively minimal data, it is a common deep learning method. In the realm of computer vision, this is incredibly helpful, particularly when the dataset is small.

**The mechanism of action and its advantages are as follows:**

* **The mechanism of action:** Transfer learning typically entails optimizing a pre-trained modela model that has been trained on a largebenchmark dataset, such as ImageNetfor a particular job. In CNNs, for instance, the deeper layers catch more specialized information, whereas the first levels capture universal elements like edges and textures. As a result, in transfer learning, the deeper layers are trained to respond to the new job whereas the earlier layers are frequently left alone or just minimally modified.

* **Advantages:** Since it makes use of previously discovered features and patterns, this method can drastically cut down on the amount of computational resources and time needed to train a deep-learningmodel. It also eliminates the need for a substantial volume of training data for the new assignment.


Finally, CNNs, data augmentation, and transfer learning are critical components of today's image-related machine learning toolset, allowing for more efficient and effective model training and deployment.

## **Task 1: Building and training a custom CNN network for AHCR:**

The CNN network was built for AHCR, where CNN Deep Learning-Based Child Handwritten Arabic Character Recognition and Handwriting Discrimination was used, and it was developed in terms of zero-padding input images, making Stride = 2 in the fourth layer, and then adding ReLU before the Fully Connected Layers. Also among the Fully Connected Layers, we chose this architecture for its simplicity, as we considered accuracy with the networks complexity, and we found that the network is effective in Arabic Handwritten Character Recognition (AHCR).


* ### **Arabic Character Recognition Model Class:**

We built the special class in the model, where the special architecture was built in the CNN model, and the function was chosen in the learning process for the parameters, and it is clear in the code the division of layers and the number of filters used.

```python
# Define your CNN model
class ArabicCharacterRecognitionModel(nn.Module):
    def __init__(self, num_classes):
        super(ArabicCharacterRecognitionModel, self).__init__()
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=0)
        self.maxpool1 = nn.MaxPool2d(kernel_size=2, stride=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.maxpool2 = nn.MaxPool2d(kernel_size=2, stride=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.maxpool3 = nn.MaxPool2d(kernel_size=2, stride=1)
        self.conv4 = nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1)
        self.maxpool4 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.relu = nn.ReLU()
        self.flatten = nn.Flatten()

        # Dropout layers
        self.dropout1 = nn.Dropout(0.2)
        self.dropout2 = nn.Dropout(0.4)

        # Fully connected layers
        self.layer1 = nn.Linear(6272, 512)
        self.layer2 = nn.Linear(512, num_classes)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu(x)
        x = self.maxpool1(x)
        x = self.conv2(x)
        x = self.relu(x)
        x = self.maxpool2(x)
        x = self.conv3(x)
        x = self.relu(x)
        x = self.maxpool3(x)
        x = self.conv4(x)
        x = self.relu(x)
        x = self.maxpool4(x)
        x = self.dropout1(x)
        x = self.flatten(x)  # Flatten the tensor
        x = self.layer1(x)
        x = self.relu(x)
        x = self.dropout2(x)
        x = self.layer2(x)
        return x
```

* ### **Custom dataset class:**

We have built a class that prepares the data stored in CSV files so that it is prepared to be entered into the model as a 2D image.

```python
# Custom dataset class
class CustomDataset(Dataset):
    def __init__(self, csv_images, csv_labels, transform=None):
        self.images = pd.read_csv(csv_images, header=None).values.astype('float32')
        self.labels = pd.read_csv(csv_labels, header=None).values.flatten().astype('int64') - 1  # Adjust labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx].reshape(32, 32, 1)/255.0 
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label
```

* This line of code was used to operate GPUs, which are useful for learning processes as they operate in parallel

```python
# Set your device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

* In these lines, the transform has been prepared and the files have been prepared for training and testing
The data was divided into batches, and this is very useful in the process of learning parameters

```python
# Load and preprocess your dataset
transform = transforms.Compose([transforms.ToTensor(),])

train_dataset = CustomDataset(csv_images='/kaggle/input/data-set/Project data set/csvTrainImages 13440x1024.csv',
                               csv_labels='/kaggle/input/data-set/Project data set/csvTrainLabel 13440x1.csv',
                               transform=transform)
test_dataset = CustomDataset(csv_images='/kaggle/input/data-set/Project data set/csvTestImages 3360x1024.csv',
                              csv_labels='/kaggle/input/data-set/Project data set/csvTestLabel 3360x1.csv',
                              transform=transform)

train_loader = DataLoader(dataset=train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=32, shuffle=False)
```

 * In this part, the model was prepared and the number of classes, which numbered 28, was taken and the loss function and optimizer were prepared, where Cross Entropy Loss was used as the loss function and Adam as the optimizer, which is considered one of the best optimizers.

```python
# Initialize your model
num_classes = len(set(train_dataset.labels))
model = ArabicCharacterRecognitionModel(num_classes=num_classes).to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

* ### **Model training and testing:**

In this part, we made 30 epochs, where we find the system reaches collision, and a report is printed for Evaluation Metrics to evaluate the system.

```python
# Train your model
num_epochs = 30
train_loss = []
val_loss = []
train_accuracy = []
test_accuracy = []
summary(model, (1, 32, 32))
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

        _, predicted = torch.max(outputs.data, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()

    epoch_loss = running_loss / len(train_loader)
    train_loss.append(epoch_loss)
    train_accuracy.append(correct_train / total_train)

    # Validation
    model.eval()
    running_val_loss = 0.0
    correct_val = 0
    total_val = 0

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            val_loss_batch = criterion(outputs, labels)
            running_val_loss += val_loss_batch.item()

            _, predicted_val = torch.max(outputs.data, 1)
            total_val += labels.size(0)
            correct_val += (predicted_val == labels).sum().item()

    epoch_val_loss = running_val_loss / len(test_loader)
    val_loss.append(epoch_val_loss)
    test_accuracy.append(correct_val / total_val)
    
       # Print classification report
    model.eval()
    all_predictions = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            
    adjusted_labels = np.array(all_labels) + 1
    adjusted_predictions = np.array(all_predictions) + 1
    report = classification_report(adjusted_labels, adjusted_predictions)

    # Calculate precision, recall, and F1-score for each class

    print(f"Epoch [{epoch+1}/{num_epochs}], "
          f"Train Loss: {epoch_loss:.4f}, "
          f"Val Loss: {epoch_val_loss:.4f}, "
          f"Train Acc: {correct_train / total_train:.4f}, "
          f"Val Acc: {correct_val / total_val:.4f}")
    
print("Classification Report:\n", report)
```

In the final report, the Precision, recall, and F1-score for each letter appears, and what concerns us is that the Precision for each letter is considered high, and this is wonderful evidence of the efficiency of the network that we built.**

## **Task 2: Retrain the network selected from Task 1 after doing data augmentation:**

In this task, the data augmentation technique was used, which is useful in increasing the diversity of data from the same data, as some changes are made to it, which increases the diversity in the data.

-> **In this cell, we created functions that prepare data augmentation for images and store it for other upcoming tasks, where 3 types of data augmentations were used, which are the following:**

* **The first is random rotation:** small degrees of rotation can mimic natural variations in handwriting, as the angle can change from person to person.

* **The second is Width and height shifts:** Changing the width and height can help simulate different handwriting sizes and the effect of connecting letters, and this produces more realistic images of the things that can happen while writing letters.

* **The third process is expansion:** This morphological process can simulate the effect of different writing pressures and the thickness of the pen due to the difference in the pens used because each pen has a different thickness from the other.

```python
def augment_dilation(image, kernel_size=3):
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    image_np = np.array(image)
    dilated = cv2.dilate(image_np, kernel, iterations=1)
    return Image.fromarray(dilated)


def augment_images(csv_images, csv_labels, output_csv_path1,output_csv_path2):
    images = pd.read_csv(csv_images, header=None).values.astype('float32')
    labels = pd.read_csv(csv_labels, header=None).values.flatten().astype('int64') - 1  # Adjust labels

    num_images = len(images)
    
    
    # Define the transformation
    rotated_image  = transforms.RandomRotation(10) 
    shifted_image  = transforms.RandomAffine(degrees=0, translate=(0.1, 0.1))
    # Placeholder for augmented images
    augmented_images = []
    augmented_labels = []

    for idx in range(num_images):
        image = images[idx].reshape(32, 32)
        label = labels[idx]
        augmented_images.append(image)
        augmented_labels.append(label)
        image = Image.fromarray(image)  # Convert to PIL Image

        # Apply transformations and store each augmented image
        for transform in [rotated_image, shifted_image, augment_dilation]:
            
            transformed_image = transform(image)
            transformed_image_array = np.array(transformed_image).reshape(32, 32)  # Reshape to 32x32
            augmented_images.append(transformed_image_array)
            augmented_labels.append(label)

    # Convert to numpy array
    augmented_images_np = np.array(augmented_images)
    augmented_label_images_np = np.array(augmented_labels)
    for i in range(len(augmented_label_images_np)):
        augmented_label_images_np[i] += 1

    # Save to CSV
    df = pd.DataFrame(augmented_images_np.reshape(num_images * 4, -1))  
    df.to_csv(output_csv_path1, index=False, header=None)
    df = pd.DataFrame(augmented_label_images_np.reshape(num_images * 4,-1))  
    df.to_csv(output_csv_path2, index=False, header=None)
# Load and preprocess your dataset

csv_TrainImages = '/kaggle/input/data-set/Project data set/csvTrainImages 13440x1024.csv'
csv_TrainLabels = '/kaggle/input/data-set/Project data set/csvTrainLabel 13440x1.csv'
csv_TrainImagesAug = '/kaggle/working/csvTrainImages 53760x1024Aug.csv'
csv_TrainLabelsAug = '/kaggle/working/csvTrainLabel 53760x1Aug.csv'
augment_images(csv_TrainImages, csv_TrainLabels,csv_TrainImagesAug,csv_TrainLabelsAug)
```

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '1'
import tensorflow as tf
# Load data
train_images = pd.read_csv('/kaggle/working/csvTrainImages 53760x1024Aug.csv').values
train_labels = pd.read_csv('/kaggle/working/csvTrainLabel 53760x1Aug.csv').values

test_images = pd.read_csv('/kaggle/input/data-set/Project data set/csvTestImages 3360x1024.csv').values
test_labels = pd.read_csv('/kaggle/input/data-set/Project data set/csvTestLabel 3360x1.csv').values

# Reshape and normalize image data
train_images = train_images.reshape(-1, 32, 32, 1).astype('float32') / 255.0
test_images = test_images.reshape(-1, 32, 32, 1).astype('float32') / 255.0

# Convert labels to one-hot encoding
encoder = OneHotEncoder(sparse_output=False)
train_labels = encoder.fit_transform(train_labels)
test_labels = encoder.transform(test_labels)

def create_lenet5_model():
    model = models.Sequential()
    model.add(layers.Conv2D(6, kernel_size=(5, 5), activation='relu', input_shape=(32, 32, 1)))
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Conv2D(16, kernel_size=(5, 5), activation='relu'))
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dense(120, activation='relu'))
    model.add(layers.Dense(84, activation='relu'))
    model.add(layers.Dense(train_labels.shape[1], activation='softmax'))  # Number of classes
    return model



# Create the optimizer with the specified learning rate
adam_optimizer = Adam(learning_rate=0.001)

model = create_lenet5_model()
model.compile(optimizer=adam_optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

history = model.fit(train_images, train_labels, epochs=30, validation_data=(test_images, test_labels), batch_size=32)
# Evaluate on validation set
val_loss, val_acc = model.evaluate(test_images, test_labels)
print(f"Validation Accuracy: {val_acc}, Validation Loss: {val_loss}")
```
